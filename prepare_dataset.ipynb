```python
''' Filter to get list of all restaurants named 'Silver Spoon' '''
regex_query_silver = { "$and":[
    {"categories" : {"$regex" : "Restaurant"} },
    {"$or": [
        {"categories" : {"$regex" : "Indian"} },
        {"categories" : {"$regex" : "Pakistani"} }#,
        #{"categories" : {"$regex" : "Korean"} }
    ]},
    {"name" : {"$regex" : "Silver Spoon"} }
]}

''' Filtered restaurant (Indian, Pakista) '''
regex_query = { "$and":[
    {"categories" : {"$regex" : "Restaurant"} },
    {"$or": [
        {"categories" : {"$regex" : "Indian"} },
        {"categories" : {"$regex" : "Pakistani"} }#,
        #{"categories" : {"$regex" : "Korean"} }
    ]},
    {"review_count": { "$gt": 10 } }
]}

''' get filterd Silver Spoon restaurants '''
filter_rest_silver = business_collection.find(regex_query_silver)
print("Total Filter Silver spoon businesses: ", filter_rest_silver.count())

''' get filterd restaurants '''
''' ** limit 96 because there are 4 restaurants named 'Silver Spoon' '''
filter_rest = business_collection.find(regex_query).limit(16)
print("Total Filter businesses: ", filter_rest.count())

''' Concate both filter results '''
restaurant_list = list(filter_rest_silver)
restaurant_list.extend(list(filter_rest))
print("Total Filter businesses: ", len(restaurant_list))
```

```python
''' process filtered restaurants to get dataset '''
def make_dataset(restaurant_list):
    restaurants = []
    all_users = []
    all_reviews = {}
    all_attrs = set()

    rest_index=0
    rest_count = len(restaurant_list)
    for record in restaurant_list:
        rest_index += 1
        print("Processing Restaurant "+str(rest_index)+"/"+str(rest_count))

        ''' making restaurant object '''
        restaurant = {"rest_id": record["business_id"], "name": record["name"], "address": record["address"],
                      "postal_code": record["postal_code"], "city": record["city"], "state": record["state"],
                      "category": record["categories"]}
        ### getting country by lat lng ###
        latitude = record["latitude"]
        longitude = record["longitude"]
        location = geolocator.reverse(str(latitude)+","+str(longitude))
        address = location.raw['address']
        country = address.get('country', '')
        restaurant["country"] = country

        ######
        restaurant["rating"] = record["stars"]

        ''' getting all attribute list of the restaurant to make Aspects '''
        attrs = []
        if record["attributes"]:
            for key, value in record["attributes"].items():
                attrs.append(key)

            ''' update all_attrs set() for unique attribute objects '''
            all_attrs.update(attrs)

        ''' fetch all photos of the restaurant to add in photos list '''
        restaurant_photos = photos_collection.find({'business_id': restaurant["rest_id"]})

        ''' process photo object'''
        photos = []
        for restaurant_photo in restaurant_photos:
            photo = {'photo_id': restaurant_photo['photo_id'], 'caption': restaurant_photo['caption'],
                     'label': restaurant_photo['label']}
            photos.append(photo)

        ''' add photos in restaurant object '''
        restaurant["photos"] = photos

        ''' fetch all reviews of the restaurant to add in review list '''
        restaurant_reviews = review_collection.find({'business_id': restaurant["rest_id"]})

        ''' process review object '''
        reviews = []

        # uncomment the following lines to limit the reviews",
        review_limit = 50
        review_count = 0
        for restaurant_review in restaurant_reviews:
            # maximum 50 reviews to be select
            review_count += 1
            if review_count > 50:
               break

            review = {"review_id": restaurant_review['review_id'], "user_id": restaurant_review['user_id'],
                      "date": restaurant_review['date'], "rating": restaurant_review['stars'],
                      "text": restaurant_review['text']}

            ''' get reviewer object from user table to get the name of reviewer and make user list '''
            user = {}
            user_obj = user_collection.find({'user_id': review["user_id"]})

            ''' set reviewer name in review object '''
            review["name"] = user_obj[0]['name']
            ''' add review in review list '''
            reviews.append(review)

            ''' make user object '''
            user['user_id'] = user_obj[0]['user_id']
            user['name'] = user_obj[0]['name']
            user['fans'] = user_obj[0]['fans']
            user['review_count'] = user_obj[0]['review_count']
            user['average_stars'] = user_obj[0]['average_stars']

            ''' make list of user object and add in user list '''
            user_friends = user_obj[0]['friends'].split(',')
            user['friends'] = user_friends
            all_users.append(user)

        ''' add reviews in restaurant object '''
        # disable append reviews to restaurant and make a separate array
        restaurant["reviews"] = reviews
        all_reviews[restaurant["rest_id"]] = reviews

        # print(restaurant['category'])
        ''' add restaurant in restaurant list '''
        restaurants.append(restaurant)

    print("Processing User data...")
    ''' make unique user list '''
    users = list({user['user_id']:user for user in all_users}.values())
    ''' remove user friends that are not in users list '''
    user_ids = set()
    for user in users:
        user_ids.add(user["user_id"])

    for user in users:
        friends_exist_arr = list(filter(lambda x: x in user_ids, user["friends"]))
        user["friends"] = friends_exist_arr

    print("Dataset prepared successfully.")
    return restaurants, users, all_reviews, all_attrs
```

```python
'''
Adding user age and gender

The Age is randomly initialized from [1, 18, 25, 35, 45, 56]
The Gender is obtained from gender prediction library
'''

import json

dataset = r"data/input/restaurant_dataset.json"
user_dataset = r"data/input/users.json"
file = open(user_dataset,)
user_data = json.load(file)
file.close()

# user = user_data[0]
# print(user)
```

```python
import random

import gender_guesser.detector as global_gender_predictor
from module.gender.guess_gender import  IndianGenderPredictor

genders = ["male", "female", "mostly_male", "mostly_female"]
age_group = [1, 18, 25, 35, 45, 56]
weights =   [1, 2, 3, 4, 3, 2]

global_gender = global_gender_predictor.Detector()
indian_gender = IndianGenderPredictor()

i = 0
for user in user_data:

    i+=1
    # if i==30:
    #     break

    # guessing user age
    name = user["name"]
    gender = global_gender.get_gender(name)
    if gender not in genders:
        gender = indian_gender.predict(name=name)
        if gender not in genders:
            user["gender"] = ""

    if gender == "mostly_male":
        gender = "male"
    if gender == "mostly_female":
        gender = "female"

    user["gender"] = gender

    # randomly guessing user age
    age = random.choices(age_group, k=1, weights=weights)[0]
    user["age"] = age

    print(str(i-1)+" name: "+name +", gender: "+gender+", age: "+str(age))
```
